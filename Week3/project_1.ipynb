```python
def reflection(img, name):
    rows, cols = img.shape
    M = np.float32([[1, 0, 0], [0, -1, rows], [0, 0, 1]])
    reflected_img = cv.warpPerspective(img, M, (int(cols), int(rows)))
    
    # Plot the reflected image
    plt.subplot(1, 2, 2)
    plt.title("Reflected Image")
    plt.imshow(reflected_img, cmap='gray')

    cv.imwrite(f'img/reflection_out_of_{name}.jpg', reflected_img)
    cv.waitKey(0)
    cv.destroyAllWindows()
    
def rotate(img, name):
    rows, cols = img.shape
    
    degree = float(input("Enter the rotation degree: "))
    # Rotate the image
    img_rotation = cv.warpAffine(
        img,
        cv.getRotationMatrix2D((cols / 2, rows / 2), degree, 0.6),
        (cols, rows)
    )
    
    # Plot the rotated image
    plt.subplot(1, 2, 2)
    plt.title("Rotated Image")
    plt.imshow(img_rotation, cmap='gray')

    # cv.imshow('img', img_rotation)
    cv.imwrite(f'img/rotation_out_of_{name}.jpg', img_rotation)
    cv.waitKey(0)
    cv.destroyAllWindows()
    
def crop(img, name):
    rows, cols = img.shape
    
    # Ask user for cropping coordinates
    x_start = int(input(f"Enter starting x-coordinate (0 to {cols-1}): "))
    x_end = int(input(f"Enter ending x-coordinate ({x_start+1} to {cols}): "))
    y_start = int(input(f"Enter starting y-coordinate (0 to {rows-1}): "))
    y_end = int(input(f"Enter ending y-coordinate ({y_start+1} to {rows}): "))

    # Ensure valid cropping range
    if x_start < 0 or x_end > cols or y_start < 0 or y_end > rows or x_start >= x_end or y_start >= y_end:
        print("Invalid cropping coordinates. Please try again.")
        exit()

    # Crop the image
    cropped_img = img[y_start:y_end, x_start:x_end]
    
    # Plot the cropped image
    plt.subplot(1, 2, 2)
    plt.title("Cropped Image")
    plt.imshow(cropped_img, cmap='gray')

    cv.imwrite(f'img/cropped_out_of_{name}.jpg', cropped_img)
    cv.waitKey(0)
    cv.destroyAllWindows()
    
def shear(img, name):
    rows, cols = img.shape
    
    axis = str(input("What axis do you want it sheared X or Y axis: "))
    axis = axis.upper()
    
    if axis != "X" and axis != "Y":
        print("Invalid axis")
        exit()
    
    if axis == "X":    
        # Shear transformation matrix
        M = np.float32([[1, 0.5, 0], [0, 1, 0], [0, 0, 1]])
        sheared_img = cv.warpPerspective(img, M, (int(cols * 1.5), int(rows * 1.5)))
    else:
        # Shear transformation along the Y-axis
        M = np.float32([[1, 0, 0], [0.5, 1, 0], [0, 0, 1]])
        sheared_img = cv.warpPerspective(img, M, (int(cols * 1.5), int(rows * 1.5)))
    
    # Plot the sheared image
    plt.subplot(1, 2, 2)
    plt.title(f"Sheared Image of {name} on {axis}-axis")
    plt.imshow(sheared_img, cmap='gray')

    cv.imshow(f'Sheared image of {name} on {axis}-axis', sheared_img)
    cv.waitKey(0)
    cv.destroyAllWindows()
    
def blur(img, name):
    cv.imshow(f'Original Image ({name})', img)
    cv.waitKey(0)  
    
    c = int(input("What type of blur do you want to employ:\n1. Guassian\n2. Median\n3. Bilateral"))
    if c==1:
        # Gaussian Blur
        strength = int(input("Enter blur strength (odd number, e.g., 3, 5, 7): "))
        if strength % 2 == 0:
            strength += 1
        blurred = cv.GaussianBlur(img, (strength, strength), 0)
    elif c==2:
        # Median Blur
        strength = int(input("Enter blur strength (odd number, e.g., 3, 5, 7): "))
        if strength % 2 == 0:
            strength += 1
        blurred = cv.medianBlur(img, strength)
    elif c==3:
        # Bilateral Blur
        strength = int(input("Enter blur strength (higher = stronger, e.g., 50, 75, 100): "))
        blurred = cv.bilateralFilter(img, 9, strength, 75)
    
    # cv.imshow(f'Blurred Image ({name})', blurred)
    cv.imwrite(f'img/blurred_{c}_{strength}_of_{name}.jpg', blurred)
    cv.waitKey(0)
    cv.destroyAllWindows()
```
