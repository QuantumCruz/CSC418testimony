```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
img = cv.imread('img/pau.jpg', 0)
rows, cols = img.shape
```

```python
# Plot the original image
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(img, cmap='gray')
```

```python
# Rotate the image
img_rotation = cv.warpAffine(
    img,
    cv.getRotationMatrix2D((cols / 2, rows / 2), 30, 0.6),
    (cols, rows)
)
```

```python
# Plot the rotated image
plt.subplot(1, 2, 2)
plt.title("Rotated Image")
plt.imshow(img_rotation, cmap='gray')

cv.imshow('img', img_rotation)
cv.imwrite('img/rotation_out.jpg', img_rotation)
cv.waitKey(0)
cv.destroyAllWindows()
```
