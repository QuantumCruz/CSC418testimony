```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('img/jare.jpg')
```

```python
plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(image)
```

```python
scaled_image = cv2.resize(image, None, fx=2, fy=2)
```

```python
cv2.imwrite('img/Scaled.jpg', scaled_image)
```

```python
plt.subplot(1,2,2)
plt.title("Scaled")
plt.imshow(scaled_image)
plt.show()
```
