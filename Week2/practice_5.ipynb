```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('img/jare.jpg')
```

```python
plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(image)
```

```python
filtered_image = cv2.medianBlur(image,15)
```

```python
cv2.imwrite('img/Median-Blur.jpg', filtered_image)
```

```python
plt.subplot(1,2,2)
plt.title("Median Blur")
plt.imshow(filtered_image)
plt.show()
```
