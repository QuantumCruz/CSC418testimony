```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('img/jare.jpg')
```

```python
plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(image)
```

```python
inverse_image = 255 - image
```

```python
cv2.imwrite('img/inverse_image.jpg', inverse_image)
```

```python
plt.subplot(1,2,2)
plt.title("Inverse Colour")
plt.imshow(inverse_image)
plt.show()
```
